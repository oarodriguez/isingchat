```python
def get_line(point_1, point_2):
    x1,y1 = point_1
    x2,y2 = point_2
    m = (y2-y1)/(x2-x1)
    
    x_data = np.arange(0,1,0.01)
    y_data = y1 + m*(x_data-x1)
    
    return (x_data, y_data)
    
```

```python
def eval_line(point_1, point_2,x0):
    x1,y1 = point_1
    x2,y2 = point_2
    m = (y2-y1)/(x2-x1)
    
    return y1 + m*(x0-x1)
```
